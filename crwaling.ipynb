```python
WHITE = (255, 255, 255)
with Image.open('20190304165422_16ead90819a164bac058e845343ee514_IMAG01_1.jpg') as im1:
    with Image.open('20190304165422_16ead90819a164bac058e845343ee514_IMAG01_2.jpg') as im2:
        # 이미지 2개를 세로로 이어서 붙일려고 합니다.
        width = max(im1.width, im2.width)
        height = im1.height + im2.height
        size = (width, height)
        with Image.new('RGB', size, WHITE) as canvas:
            canvas.paste(im1, box=(0, 0)) # left/top 지정
            canvas.paste(im2, box=(0, im1.height)) # left/top 지정
            canvas.save('canvas.jpg')
```

# 정리

```python
import re
import json
import requests

def naver_spellchecker(q):
    page_url = 'https://m.search.naver.com/p/csearch/ocontent/util/SpellerProxy'
    params = {
        '_callback' : '_', # 이거를 _로 바꾸면 바뀐다.
        'q' : q,
        'where' : 'nexearch',
        'color_blindness' : '0',        
    }
    res = requests.get(page_url, params=params)
    jsonp_string = res.text
    json_string=re.sub(r'(^_\(|\);$)', '', jsonp_string)
    obj = json.loads(json_string)
    return obj['message']['result']
```

```python
naver_spellchecker('아버지가방에들어가신다')
```
