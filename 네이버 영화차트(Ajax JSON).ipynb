```python
def get_naver_movie_chart(chart_type):
    chart_url = 'https://movie.naver.com/movieChartJson.nhn'
    params = dict(type=chart_type)
    
    #파리미터를 수정
    #예매순 : RESERVE
    # 현재 상영작 : CURRENT
    # 다운로드 : DOWNLOAD
    #개봉예정작 : COMMING
    # 평점순 : POINT
    # 박스오피스 : BOXOFFICE

    headers={
        'User-Agent' : ('Mozilla/5.0 (Windows NT 10.0; Win64; x64)' 
                        'AppleWebKit/537.36 (KHTML, like Gecko)' 
                        'Chrome/74.0.3729.131 Safari/537.36'),
        'Referer' : 'https://movie.naver.com',
    }
    res = requests.get(chart_url, params=params, headers = headers)    
    movie_list = res.json()['movieChartList'][chart_type]
    for rank, movie in enumerate(movie_list,1):
        template = "{} {movieCode} {movieTitle} {reserveRatio}%"
        print(template.format(rank, **movie))
```

```python
def get_movie_detail(movie_code):
    page_url = 'https://movie.naver.com/movie/bi/mi/basic.nhn'
    params = dict(code = movie_code)
    res = requests.get(page_url, params=params)

    html = res.text
    soup = BeautifulSoup(html, 'html.parser')

    tag_list = tag_list=soup.select('.mv_info_area .main_score .st_on')[:3]
    label_list = ['평점-관람객', '평점-평론가', '평점-네티즌']
    percent_list = []

    for tag in tag_list:
        matched = re.match(r'width:([\d\.]+)%', tag['style'])
        percent = '%.2f' % (float(matched.group(1))/10)
        percent_list.append(percent)
        
    detail = {}
    detail.update(dict(zip_longest(label_list, percent_list, fillvalue=0)))
    return detail

def get_naver_movie_chart(chart_type):
    chart_url = 'https://movie.naver.com/movieChartJson.nhn'
    params = dict(type=chart_type)
    
    #파리미터를 수정
    #예매순 : RESERVE
    # 현재 상영작 : CURRENT
    # 다운로드 : DOWNLOAD
    #개봉예정작 : COMMING
    # 평점순 : POINT
    # 박스오피스 : BOXOFFICE

    headers={
        'User-Agent' : ('Mozilla/5.0 (Windows NT 10.0; Win64; x64)' 
                        'AppleWebKit/537.36 (KHTML, like Gecko)' 
                        'Chrome/74.0.3729.131 Safari/537.36'),
        'Referer' : 'https://movie.naver.com',
    }
    res = requests.get(chart_url, params=params, headers = headers)    
    movie_list = res.json()['movieChartList'][chart_type]
    
    for rank, movie in enumerate(movie_list,1):
        detail = get_movie_detail(movie['movieCode'])
        movie.update(detail)
        movie['rank'] = rank
        yield movie
        time.sleep(0.1)
        
```
