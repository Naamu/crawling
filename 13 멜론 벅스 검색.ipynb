```python
headers = {
    'User-Agent' : user_agent,
}
res = requests.get(melon_chart_url, headers=headers)
res
```

```python
html = res.text
soup = BeautifulSoup(html, 'html.parser')
```

```python
tag_list = soup.select('.lst50')
len(tag_list)
```

```python
tag_list = soup.select('.lst50, .lst100') # 콤마로 이거나 지정
len(tag_list)
```

```python
tag=tag_list[0]
song_no = tag['data-song-no']
song_title = tag.select_one('[href*=playSong]').text # 태그리스트 중에서 playSong 이 포함된 것
artist_name = tag.select_one('[href*=goArtistDetail]').text
#album_name = tag.select_one('[href*=goAlbumDetail]').text  # 앞에 있는게 선택되서 안나옴.
album_name = tag.select_one('.wrap_song_info [href*=goAlbumDetail]').text

print(song_no, song_title, artist_name, album_name)
```

```python
import re 

def get_number_string(s):
    mached = re.search(r'(\d+)', s)
    return mached.group(1)
```

```python
tag=tag_list[0]

song_no = tag['data-song-no']
song_title = tag.select_one('[href*=playSong]').text # 태그리스트 중에서 playSong 이 포함된 것

artist_tag = tag.select_one('[href*=goArtistDetail]')
artist_name = artist_tag.text

artist_no = get_number_string(artist_tag['href'])

#album_name = tag.select_one('[href*=goAlbumDetail]').text  # 앞에 있는게 선택되서 안나옴.
album_tag = tag.select_one('.wrap_song_info [href*=goAlbumDetail]')
album_name = album_tag.text

album_no = get_number_string(album_tag['href']) # 왜인지는 알수없지만 앨범 숫자가 안나옴. 오탈



{
    'song_no' : song_no,
    'song_title' : song_title,
    'artist_no' : artist_no,
    'artist_name' : artist_name,
    'album_no' : album_no,
    'album_name' : album_name,
    
}
```

```python
song_list = []
for tag in tag_list : 
    song_no = tag['data-song-no']
    song_title = tag.select_one('[href*=playSong]').text
    artist_tag = tag.select_one('[href*=goArtistDetail]')
    artist_no = get_number_string(artist_tag['href'])
    artist_name = artist_tag.text    

    album_tag = tag.select_one('.wrap_song_info [href*=goAlbumDetail]')
    album_no = get_number_string(album_tag['href'])
    album_name = album_tag.text
    
    song_list.append({
        'song_no' : song_no,
        'song_title' : song_title,
        'artist_no' : artist_no,
        'artist_name' : artist_name,
        'album_no' : album_no,
        'album_name' : album_name,
    })
song_list
```

```python
import pandas as pd
pd.DataFrame(song_list).set_index('song_no').to_excel('melon_chart.xlsx')
```
